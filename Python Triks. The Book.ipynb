In both cases, using a with statement allows you to abstract away most
of the resource handling logic. Instead of having to write an explicit
try...finally statement each time, using the with statement takes
care of that for us.

````python
class MyContextManager:
    def __init__(self):
        print("init")
    def __enter__(self):  
        print("enter")
    def __exit__(self, exc_type, exc_value, exc_traceback):
        print("exit")
````

````python
with MyContextManager() as manager:
    print("do")
````

What’s a context manager? It’s a simple “protocol” (or interface) that
your object needs to follow in order to support the with statement.
Basically, all you need to do is add __enter__ and __exit__ methods
to an object if you want it to function as a context manager. Python
will call these two methods at the appropriate times in the resource
management cycle

````python
# Explore 

class FileManager():
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None
          
    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file
      
    def __exit__(self, exc_type, exc_value, exc_traceback):
        self.file.close()

# loading a file 
with FileManager('test.txt', 'w') as f:
    f.write('Test')

print(f.closed)
````


# 3.3 The Power of Decorators

````python
# The function’s behavior changes only when it’s decorated.

# a decorator is a callable that takes a callable as input and
# returns another callable
````

````python
def null_decorator(fn):
    return fn
````

````python
def greet():
    return 'Hello'
````

````python
greet = null_decorator(greet)
````

````python
greet()
````

````python
@null_decorator
def yell():
    return "Yellll"
````

````python
yell()
````

### Decorators Can Modify Behavior

````python
def uppercase(fn):
    def wrapper():
        original_result = fn()
        modified_result = original_result.upper()
        
        return modified_result
    
    return wrapper
    
````

````python
@uppercase
def yell():
    return "Yellll"
````

````python
yell()
````

### Applying Multiple Decorators to a Function

````python
def strong(fn):
    def wrapper():
        return  '<b>' + fn() + '</b>'
    
    return wrapper
````

````python
def emp(fn):
    def wrapper():
        return  '<em>' + fn() + '</em>'
    
    return wrapper
````

````python
@emp
@strong
def yell():
    return "Yellll"
````

````python
yell()
````

### Decorating Functions That Accept Arguments

````python
def proxy(fn):
    def wrapper(*args):
        return 'Hello ' + args[0]
    
    return wrapper
````

````python
@proxy
def meet(name):
    return name
````

````python
meet("valod")
````

````python
import functools
````

````python
def uppercase(fn):
    @functools.wraps(fn)
    def wrapper():
        return fn().upper()
    
    return wrapper
````

````python
@uppercase
def greet():
    """Return a friendly greeting."""
    return 'hello'
````
