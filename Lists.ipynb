# append(), extend(), insert()


### 1. usage

```python
a = [1,2,3,4,5]
```

```python
a.append(9)
a
```

```python
# append as list not as elements
a.append([7,7,8])
a
```

```python
a.extend([7,7,8])
a
```

```python
# also sets, tuples
a.extend({90, 91})
a.extend((92, 93))
a
```

```python
a.insert(1,88)
a
```

### 2. convert to each other

```python
# append -> insert (not work with -1)
a.insert(-1, 44)
a
```

```python
# append -> insert 
a.insert(len(a), 99)
a
```

```python
# append -> extend 
a.extend([100])
a
```

```python
# append -> list slicing syntax
a[len(a):] = [22]
a
```

```python
# extend -> list slicing syntax
a[len(a):] = [55,66,77]
a
```

```python
# insert -> list slicing syntax
a = a[:2] + [55555] + a[2:]
a
```

### 3. return value is NoneType 

```python
a.append(11).append(12)
```

```python
a.extend([1,2]).append(4)
```

```python
a.insert(0,1).append(0)
```

```python
v = [1,2,3]
v
```

```python
v.clear()
v
```

### 2. convertions

```python
a
```

```python
# pop() -> list slice syntax
a = a[:-1]
a
```

```python
# pop(i) -> list slice syntax
a = a[:2] + a[3:]
a
```

```python
# pop(i) -> del 
del a[1]
a
```

```python
# remove(x) -> pop()
a.pop(a.index(9))
a
```

```python
# clear -> del a[:]
del a[:]
a
```

# index(), count(), copy()

### 1. usage


```python
a = [4,8,3,9,65,11,4,3,55]
```

```python
a.index(4)
```

```python
# get next index besides the first
a.index(4, 1)
```

```python
# auto detect 2 appearance
a.index(4, a.index(4)+1)
```

```python
# last appearance
len(a) - a[::-1].index(4) - 1
```

```python
a
```

```python
a.count(4)
```

```python
# shallow
c = a.copy()
c
```

### 2. convertions

```python
# copy -> list syntax
m = a[:]
m
```
